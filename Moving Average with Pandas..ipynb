```python
import pandas as pd
from datetime import datetime
pd.core.common.is_list_like = pd.api.types.is_list_like # is_list_like is moved to api in pandas, but pandas_datareader expects this to be in pd.core.common
from pandas_datareader import data


ticker = 'AAPL'
start_date =  datetime(2013, 10, 1)
end_date = datetime(2018, 4, 30)

panel_data = data.DataReader(ticker, 'iex', start_date, end_date)
close = panel_data['close']
close = pd.DataFrame(close, columns=['close'])
close.to_csv('aapl_closing_stock_price.csv')
moving_average = pd.read_csv('aapl_closing_stock_price.csv')
moving_average['date'] = pd.to_datetime(moving_average['date'])
moving_average['date'] = pd.to_numeric(moving_average['date'])
moving_average.rolling(window=5).mean()
```
